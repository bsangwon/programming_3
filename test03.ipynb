## usecols=cols를 사용하여 특정 컬럼만 선택

위와 같은 Dtype변경 후에도 메모리가 모자라면 특정 컬럼만 선택하는 방법이 있습니다.

특정 컬럼을 선택해서 내가 필요없는 부분을 제외하고 불러오는 방법인

read.csv의 usecols를 사용해서 cols에 내가 원하는 컬럼들의 이름을 정해주면 그 컬럼들만 선택해서 가져옵니다.

```python
#cols에서  'movie_imdb_link', 'plot_keywords'을 제외
cols = ['color', 'director_name', 'num_critic_for_reviews', 'duration', 'director_facebook_likes', 'actor_3_facebook_likes', 'actor_2_name', 
        'actor_1_facebook_likes', 'gross', 'genres', 'actor_1_name', 'movie_title', 'num_voted_users', 'cast_total_facebook_likes',
        'actor_3_name', 'facenumber_in_poster', 'num_user_for_reviews', 'language', 'country',
        'content_rating', 'budget', 'title_year', 'actor_2_facebook_likes', 'imdb_score', 'aspect_ratio', 'movie_facebook_likes']

movies4 = pd.read_csv('data/movie.csv', nrows=4000,
                     dtype={
                         'num_critic_for_reviews':np.float16,
                         'duration':np.float16,
                         'director_facebook_likes':np.float16,
                         'actor_3_facebook_likes':np.float16,
                         'actor_1_facebook_likes':np.float32,
                         'gross':np.float32,
                         'facenumber_in_poster':np.float16,
                         'num_user_for_reviews':np.float16,
                         'budget':np.float32,
                         'title_year':np.float16,
                         'actor_2_facebook_likes':np.float32,
                         'imdb_score':np.float16,
                         'aspect_ratio':np.float16,
                         'num_voted_users':np.int32,
                         'cast_total_facebook_likes':np.int32,
                         'movie_facebook_likes':np.int32,
                         'color':'category',
                         'language':'category',
                         'country':'category',
                         'content_rating':'category',
                     },
                     usecols=cols)
movies4.info()
```

```python
round(387.5/875.1, 2)
```

cols에서 임의로 'movie_imdb_link', 'plot_keywords'컬럼을 제외한 경우 처음 데이터 875.1KB에서 현재 387.5KB로 처음과 비교해 44%의 메모리만 사용합니다.

## chunksize 사용

만약 이렇게 했는데도 공간이 모자라면 데이터를 나눠서 로딩하는 방법이 있습니다.

read_csv 에서 chunksize를 설정해 데이터를 한번에 몇 줄을 불러올 것인지를 정합니다. 

```python
cols = ['color', 'director_name', 'num_critic_for_reviews', 'duration', 'director_facebook_likes', 'actor_3_facebook_likes', 'actor_2_name', 
        'actor_1_facebook_likes', 'gross', 'genres', 'actor_1_name', 'movie_title', 'num_voted_users', 'cast_total_facebook_likes',
        'actor_3_name', 'facenumber_in_poster', 'num_user_for_reviews', 'language', 'country',
        'content_rating', 'budget', 'title_year', 'actor_2_facebook_likes', 'imdb_score', 'aspect_ratio', 'movie_facebook_likes']

movies_iter = pd.read_csv('data/movie.csv', nrows=4000,
                     dtype={
                         'num_critic_for_reviews':np.float16,
                         'duration':np.float16,
                         'director_facebook_likes':np.float16,
                         'actor_3_facebook_likes':np.float16,
                         'actor_1_facebook_likes':np.float32,
                         'gross':np.float32,
                         'facenumber_in_poster':np.float16,
                         'num_user_for_reviews':np.float16,
                         'budget':np.float32,
                         'title_year':np.float16,
                         'actor_2_facebook_likes':np.float32,
                         'imdb_score':np.float16,
                         'aspect_ratio':np.float16,
                         'num_voted_users':np.int32,
                         'cast_total_facebook_likes':np.int32,
                         'movie_facebook_likes':np.int32,
                         'color':'category',
                         'language':'category',
                         'country':'category',
                         'content_rating':'category',
                     },
                     usecols=cols,
                     chunksize=500)

def process(df):
    return f'processed {df.shape} shape of df' #{df.iloc[0, :]}

for chunk in movies_iter:
    print(process(chunk))
```

4000개의 row를 500개씩 나눠서 불러온 것을 알 수 있습니다.
